```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np

rcParams['font.weight'] = 'bold'
plt.rcParams.update({'font.size': 24})
```

```python
Nt0 = 180 #Dividing n theta into 180 grid of 180
Nr = 250 #Radial direction as grid of Nr

X1 = np.linspace(0,0.55,Nr)
delt = np.pi*(2/(1+np.exp(-8*X1**(1.32)))-1)

t0_grid = np.linspace(0,2*np.pi,Nt0)
dt0 = t0_grid[1]-t0_grid[0]

X = np.linspace(0,1,Nr)
```

```python
I1 = np.loadtxt("AllSkyPC12.csv", delimiter=",")*250/2
I2 = np.loadtxt("AllSkyAPC12.csv", delimiter=",")*250/2
I3 = np.loadtxt("AllSkyPC6.csv", delimiter=",")*250/2
I4 = np.loadtxt("AllSkyAPC6.csv", delimiter=",")*250/2
```

```python
print(I1[0],I2[0],-I1[0]+I2[0])
print(np.max(I1),np.max(I2),np.max(I1)-np.max(I2))
```

```python
# Load your data
#I_anti = np.loadtxt("I_par.csv")
#I_par = np.loadtxt("I_anti.csv")

x = X
y2 = I1
y1 = I3

# Configure font settings
plt.rcParams['font.weight'] = 'bold'
plt.rcParams.update({'font.size': 30})

# Create a figure with a white background
fig = plt.figure(figsize=(10, 8))
fig.patch.set_facecolor('white')

# Plot the data and set line properties
cp, = plt.plot(x, y1, label='P', color='b', linewidth=3)
cn, = plt.plot(x, y2, label='AP', color='r', linewidth=3, zorder=3)

# Set labels and title
plt.xlabel('Radial Distance/Radius', fontweight='bold', fontsize=36)
plt.ylabel('Current Density (A/m²)', fontweight='bold', fontsize=36)

# Set custom axis limits and tick positions
x_min, x_max = 0,1
y_min, y_max = 0,8*1e9
x_ticks = [0, 0, 0.1, 0.2]
y_ticks = [2*1e9, 4*1e9, 6*1e9,8*1e9]

#plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.xlim(x_min, x_max)
#plt.ylim(y_min, y_max)

# Create a legend without a box outline
lgd = plt.legend([cp, cn], ['6nm', '12nm'])
lgd.set_frame_on(False)

# Customize tick parameters and spines
plt.gca().tick_params(axis='both', which='major', width=3)
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)

plt.tight_layout()
plt.savefig('ASP.png')

# Show the plot
plt.show()
```
